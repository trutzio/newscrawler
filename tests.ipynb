```python
import boto3

s3 = boto3.client('s3', region_name='us-east-1')
warc_files = [file['Key'] for file in s3.list_objects_v2(Bucket='commoncrawl', Prefix='crawl-data/CC-NEWS/2016/08').get('Contents', []) if file['Key'].endswith('.warc.gz')] 
warc_files
 
```

```python
warc_file_path = 'crawl-data/CC-NEWS/2016/08/CC-NEWS-20160826124520-00000.warc.gz'
warc_file_name = warc_file_path.split('/')[-1]
with open(warc_file_name, 'wb') as f:
    s3.download_fileobj('commoncrawl', warc_file_path, f)

```

```python
from warcio.archiveiterator import ArchiveIterator
import gzip

with gzip.open(warc_file_name, 'rb') as stream:
    for record in ArchiveIterator(stream):
        if record.rec_type == 'response':
            url = record.rec_headers.get_header('WARC-Target-URI')
            content_type = record.http_headers.get_header('Content-Type')
            print(f'URL: {url}, Content-Type: {content_type}')
```
